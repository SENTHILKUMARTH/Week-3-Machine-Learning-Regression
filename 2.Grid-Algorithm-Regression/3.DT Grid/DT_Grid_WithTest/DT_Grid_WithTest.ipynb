```python
import pandas as pd
```

```python
dataset = pd.read_csv("insurance_pre.csv")
```

```python
dataset = pd.get_dummies(dataset,dtype=int, drop_first=True)
```

```python
dataset.columns
```

```python
independent = dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dependent = dataset[['charges']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size=0.30, random_state=True)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
param_grid= {
            'criterion': ['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
            'splitter': ['best', 'random'],
            'max_features':['sqrt', 'log2']
            }
grid = GridSearchCV(DecisionTreeRegressor(), param_grid, n_jobs=-1,refit=True,cv=5, verbose=3)
grid.fit(X_train, y_train)
```

```python
grid_pred = grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score = r2_score(y_test, grid_pred)
print("r_score{}".format(grid.best_params_),r_score)
```

```python
re = grid.cv_results_
table = pd.DataFrame.from_dict(re)
table
```
