```python
import pandas as pd
```

```python
dataset = pd.read_csv("insurance_pre.csv")
```

```python
dataset = pd.get_dummies(dataset, dtype=int, drop_first=True)
```

```python
dataset.columns
```

```python
independent = dataset[['age', 'bmi', 'children',  'sex_male', 'smoker_yes']]
dependent = dataset[['charges']]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV
param_grid = {
                'fit_intercept':[True, False],
                'copy_X':[True, False],
                'n_jobs': [None, 1, -1],
                'positive':[False, True]
             }
grid = GridSearchCV(LinearRegression(), param_grid, n_jobs=-1, refit=True, cv=5, verbose=3)
grid.fit(independent, dependent)
```

```python
grid_predict = grid.predict(independent)
```

```python
from sklearn.metrics import r2_score
r_score = r2_score(dependent, grid_predict)
print("r2 value {}".format(grid.best_params_),r_score)
```

```python
re = grid.cv_results_
table = pd.DataFrame.from_dict(re)
```
